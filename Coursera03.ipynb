```python
import requests # library to handle requests

LIMIT = 5
radius = 500

location_list = [] # initiate a list to store data from Foursquare API requests

for neighbourhood, latitude, longitude in zip(df.Neighbourhood, df.Latitude, df.Longitude):
    
    url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
    CLIENT_ID, 
    CLIENT_SECRET, 
    VERSION, 
    latitude, 
    longitude, 
    radius, 
    LIMIT)
    
    data = requests.get(url).json()
    # use len() to check if any data within "items" (len = 0 if nothing)
    length = data['response']['groups'][0]['items']
    if length == []:
        continue # skip the row if nothing is found
    else:
        venue = data['response']['groups'][0]['items'][0]['venue']

     # extract info within 'venue'
        name = venue['name']
        lat = venue['location']['lat']
        lon = venue['location']['lng']
        cat = venue['categories'][0]['name']

        location_list.append([(neighbourhood, latitude, longitude, name, lat, lon, cat)])
```

- making dataframe from the laste foursquare request 

```python
temp = pd.DataFrame(x for row in location_list for x in row)
temp.columns = ['Neighbourhood','Nei_Latitude','Nei_Longitude','Venue','Ve_Latitude','Ve_Longitude','category']
temp.head()
```

# making one hot encoding for K-means

```python
cat = pd.get_dummies(temp.category) # one hot encoding
df1=pd.concat([temp[['Neighbourhood']], cat], axis=1)
df1.head()
```

```python
from sklearn.cluster import KMeans
```

```python
df2 = df1.drop('Neighbourhood', axis=1)

n = 6 

# run k-means clustering
kmeans = KMeans(n_clusters=n, random_state=0).fit(df2)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]
```

```python
df1.insert(1, 'label', kmeans.labels_)
```

# merging the tow data frame and show clusters on the maps

```python
toronto = pd.merge(df, df1, on='Neighbourhood', how='right')
toronto.head()
```

```python
import matplotlib.cm as cm
import matplotlib.colors as colors

ap_clusters = folium.Map(location=[latitude, longitude], zoom_start=10)

# set color scheme for the clusters
x = np.arange(n)
ys = [i + x + (i*x)**2 for i in range(n)]
colors1 = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors1]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto['Latitude'], toronto['Longitude'], toronto['Neighborhood'], toronto['label']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_)
       
map_
```
